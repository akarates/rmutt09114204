```python
import os
import subprocess
```

```python
def convert(dirname):
    for _, _, files in os.walk(dirname):
        for f in files:
            if f.endswith('.ipynb') and 'checkpoint' not in f:
                # print(f)
                name = f[:-6]
                print(f'Convert {f} to html ...', end=' ')
                subprocess.run(['jupyter', 
                                'nbconvert', 
                                '--to', 
                                'html', 
                                dirname+'/'+f])
                print(f'Wrote {name}.html successfully')
                print('-'*45)
                
                print(f'Convert {name}.html to pdf ...', end=' ')
                chrome = r'/Applications/Google Chrome.app/Contents/MacOS/Google Chrome'
                subprocess.run([chrome, 
                                '--headless', 
                                f'--print-to-pdf={dirname}_{name}.pdf', 
                                f'{dirname}/{name}.html'])
                print(f'Wrote {dirname}_{name}.pdf successfully') 
                print('='*45)
         
```

```python
dirs = ['oop_b', 'exceptions_b', 'files_b', 'numpy_b', 'pandas_b', 'python_b']
for d in dirs:
    convert(d)
    convert(d+'/exercises')
```
